### Draw Chart on GPDB

```python
sql = """
SELECT home_project.py_basic_plot_chart(X, Y, 
    'Stock Price', 
    legend, 
    NULL, 
    opt) chart
FROM (
SELECT string_agg(s_close::text, ',' order by s_date) Y,
    string_agg((s_date - '2017-01-01'::date)::text, ',' order by s_date) X,
    string_agg(code, ',' order by s_date) legend,
    '{
    \"figure.figsize\": [12.0, 6.0], 
    \"legend.loc\":\"upper right\",
    \"lines.marker\":\"o\",
    \"axes.color_map\":\"PuOr\",
    \"axes.color_alpha\":0.7
    }' opt
FROM stock.price
WHERE code IN ('090435.KS')
    AND s_date >= '2017-11-01' AND s_date < '2017-12-01'
) as a;
"""
```

```python
cur = conn.cursor()
cur.execute(sql)
rows = cur.fetchall()
cur.close()
```
